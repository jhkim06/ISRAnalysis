```python
plot_test = Plotter.Plotter(2, 1, gridspec_kw={'height_ratios':[1,0.3]})
```

```python
plot_test.clear_all_axis()

hist_path = "[tunfold-hist]_[dipt-dimass]_[folded_extended-folded_dimass55]"
axis_steering = "dipt[];dimass[UOC0]"

file_key = "mc/signal:Drell-Yan/all"
file_key_list = ["mc/background:tau/all",
                 "mc/signal:Drell-Yan/all"]

dy_2016a_hist = ee2016a.get_hist("mc/signal:Drell-Yan/all", hist_path, axis_steering, binwnorm=True)
dy_2016a_0bject_hist = ee2016a.get_hist("mc/signal:Drell-Yan/all", hist_path, axis_steering, binwnorm=True, 
                                        file_sel="0bjet__:0bjet")


axis = plot_test.get_axis(0, 0)
hep.cms.label("Simulation", data=True, year="Run 2", ax=axis, fontsize=30, loc=0)

plot_test.draw_hist(dy_2016a_hist, histtype='errorbar', label='w/o b veto', xerr=True, 
                    yerr=dy_2016a_hist.errors, color='black', facecolor=None)
plot_test.draw_hist(dy_2016a_0bject_hist, histtype='step', label='b veto', xerr=True, 
                    color='gray', facecolor=None)


axis.set_xticklabels([])
hep.plot.hist_legend(axis, loc='best')

axis = plot_test.get_axis(1, 0)

ratio = ee2016a.get_ratio_hist([file_key], [file_key],
                               hist_path, hist_path, axis_steering, binwnorm=True,
                              file_sel2="0bjet__:0bjet")


plot_test.draw_hist(ratio, row=1, col=0, xerr=True, yerr=ratio.errors, histtype='errorbar', 
                    label='Data', color='black')

axis.set_ylim(0.6, 1.4)
axis.axhline(y=1, color='gray', linestyle='--')

plot_test.save_plot("comparison")
```

```python
plot_test.clear_all_axis()

hist_path = "[tunfold-hist]_[dipt-dimass]_[folded_extended-folded_dimass55]"
axis_steering = "dipt[];dimass[UOC4]"

file_key = "data/2016prePFV/all"
file_key_list = ["mc/background:tau/all",
                 "mc/signal:Drell-Yan/all"]

dy_2016a_hist = ee2016a.get_hist(file_key, hist_path, axis_steering, binwnorm=True)
dy_2016a_mc_hist = ee2016a.get_combined_hist(file_key_list, hist_path, axis_steering, binwnorm=True)
systematic_dict = {
    "PUweight": ["up", "down"],
    "btagSF": ["hup", "hdown"],
    "noz0weight": [""]
}

file_key = "mc/signal:Drell-Yan/all"

systematic = ee2016a.get_combined_hist_systematic_bands(systematic_dict, file_key_list, hist_path, axis_steering, 
                                                        binwnorm=True, file_sel="SYS__:")

axis = plot_test.get_axis(0, 0)
hep.cms.label("Simulation", data=True, year="Run 2", ax=axis, fontsize=30, loc=0)

plot_test.draw_hist(dy_2016a_hist, histtype='errorbar', label='Data', xerr=True, 
                    yerr=dy_2016a_hist.errors, color='black', facecolor=None)
plot_test.draw_hist(dy_2016a_mc_hist, histtype='step', label='MC total', xerr=True, 
                    color='gray', facecolor=None)
plot_test.draw_error_boxes(dy_2016a_mc_hist, sys_yerror=systematic, 
                           facecolor='none', edgecolor='black', alpha=0.2)

axis.set_xticklabels([])
hep.plot.hist_legend(axis, loc='best')

axis = plot_test.get_axis(1, 0)

ratio = ee2016a.get_ratio_hist(["data/2016prePFV/all"], file_key_list,
                               hist_path, hist_path, axis_steering, binwnorm=True)

scaled, sys = ee2016a.get_scaled_hist(file_key_list, hist_path, axis_steering, binwnorm=True, 
                                         file_sel="SYS__:", systematic_dict=systematic_dict)

plot_test.draw_hist(ratio, row=1, col=0, xerr=True, yerr=ratio.errors, histtype='errorbar', 
                    label='Data', color='black')
plot_test.draw_error_boxes(scaled, row=1, col=0, facecolor='none', edgecolor='black', alpha=0.2)

axis.set_ylim(0.6, 1.4)
axis.axhline(y=1, color='gray', linestyle='--')

plot_test.save_plot("comparison")
```
